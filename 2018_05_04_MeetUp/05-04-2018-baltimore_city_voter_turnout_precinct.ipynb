```python
import pandas as pd
import numpy as np
import folium
import geopandas as gpd
```

```python
precinct = 'https://raw.githubusercontent.com/melanieshimano/python4CityGovtProcessImprovement/master/May2018_MeetUp/Official%20by%20Party%20and%20Precinct.csv'
```

```python
df=pd.read_csv(precinct)
```

```python
df.head()
```

```python
df_bmore = df[df['LBE']=='Baltimore City']
```

```python
df_bmore.head()
```

```python
#New df that sums all votes for specific precinct
df_bmore_new= df_bmore.groupby('PRECINCT', as_index=False)[('ELIGIBLE_VOTERS', 'POLLS', 'EARLY_VOING', 'ABSENTEE', 'PROVISIONAL')].sum()
```

```python
#total number of precincts in dataset
df_bmore_new['PRECINCT'].nunique()
```

```python
#calculate total number of votes in each precinct
df_bmore_new['TOTAL_VOTES']=df_bmore_new['EARLY_VOING']+df_bmore_new['POLLS']+df_bmore_new['ABSENTEE']+df_bmore_new['PROVISIONAL']
```

```python
df_bmore_new.head()
```

```python
#calculate voter turnout for each precinct
df_bmore_new['TURNOUT']=round(((df_bmore_new['TOTAL_VOTES']/df_bmore_new['ELIGIBLE_VOTERS'])*100),2)
```

```python
df_bmore_new.sort_values(by='TURNOUT', ascending = False)
```

```python
#Baltimore City Precinct 011-004 percentage is over 100%--need to look into data
#Eligible Voters = 63
#Polls = 28
#Early Voting = 8
#Absentee= 1
#Provisional = 95
#Total Votes = 132
```

```python
#removing precinct 011-004 for clarity for now
df_bmore_new=df_bmore_new[df_bmore_new['TURNOUT']<100]
```

```python
df_bmore_new.sort_values(by='TURNOUT', ascending = False)
```

```python
#baltimore city shapefile for precincts
precinct_shp = gpd.read_file('/Users/melanieshimano/Downloads/2010 Census Wards Precincts/geo_export_8e4c4012-d8e2-4a9d-aee6-9ea5c3fd9abe.shp')
```

```python
precinct_shp.head()
```

```python
precinct_shp['name_1'].nunique()
```

```python
df_bmore_new.info()
```

```python
#edit df to match shapefile format

df_bmore_new['PRECINCT']=df_bmore_new['PRECINCT'].str[1:]
df_bmore_new['PRECINCT']='Baltimore City Precinct '+ df_bmore_new['PRECINCT']
```

```python
df_bmore_new.head()
```

```python
#merge dataset with the shapefile data
#data is the shapefile, df_bmore_new is the created dataframe to merge onto the shapefile
#right_on is from df_bmore_new, left_on is from data -- use these if the two dataframes have two different names
#if the join columns have the same name, then you can use on='join column'

join = precinct_shp.merge(df_bmore_new, right_on='PRECINCT', left_on ='name_1')
```

```python
join.head()
```

```python
#set Baltimore City coordinates
BMORE_COORD=(39.2904, -76.6122)
```

```python
#set Baltimore City map
m_16 = folium.Map(BMORE_COORD, zoom_start=11.4)
```

```python
m_16.choropleth(geo_data = join,
    data=df_bmore_new,
    columns=['PRECINCT','TURNOUT'],
    key_on= 'feature.properties.name_1',
    fill_opacity=0.7,
    line_opacity=0.2,
    fill_color='BuPu',
    legend_name='Voter Turnout for 2016 General Election in Baltimore City by Precinct'
)
```
