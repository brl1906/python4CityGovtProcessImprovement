#### conversion of fine and balance data

```python
# create function to convert fine amount data to floats
def cash_conversion(amount):
    return float(amount.replace('$','').replace(',',''))
```

```python
# apply conversion function and set column as result
dataframe['FineAmount'] = dataframe['FineAmount'].apply(cash_conversion)
dataframe['Balance'] = dataframe['Balance'].apply(cash_conversion)
```

### <font color=green>Taking a quick peak</font>

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(14,7))
ax1 = fig.add_subplot(1,2,1)
plt.plot(dataframe.resample('A')['FineAmount'].mean(),label='Avg fine',\
        lw=6,color='green',alpha=.5)
plt.plot(dataframe.resample('A')['Balance'].mean(),label='Avg balance',\
        lw=3,color='red',alpha=.5)
plt.legend()
plt.ylabel('dollars')
plt.title('Avg Fine & Avg Balance Owed')
plt.grid(False)

ax1a = ax1.twinx()
plt.plot(dataframe.resample('A')['CitationNo'].count(),\
         label='citations',ls='--',color='black')
plt.plot(dataframe[dataframe['CitationStatus'] == 'A       '].\
         resample('A')['CitationStatus'].count(),label='abated',\
        lw=6,color='orange')
plt.legend()
plt.grid(False)
plt.ylim(50,60000);
```
